```python
import numpy as np
import pandas as pd
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import OrdinalEncoder
from sklearn.impute import SimpleImputer
from sklearn.compose import ColumnTransformer
```

```python
df = pd.read_csv('Data Sets//covid_toy.csv')
df
```

1. gender, city -> nominal categorical data
2. age,fever -> numerical data
3. cough -> ordinal categorical data

Train Test Split

```python
x_train, x_test, y_train, y_test = train_test_split(df.drop(columns= 'has_covid'), df['has_covid'], test_size=0.3, random_state=0)
```

```python
x_train.shape, x_test.shape
```

# Aam Zindagi -> traditional method -> without ColumnTransformer

```python
# adding simple imputer to fever col
si = SimpleImputer()
x_train_fever = si.fit_transform(x_train[['fever']])

# also the test data
x_test_fever = si.fit_transform(x_test[['fever']])
x_train_fever.shape
```

```python
# ordinal encoding
oe = OrdinalEncoder(categories= [['Mild','Strong']])
x_train_cough = oe.fit_transform(x_train[['cough']])

# also the test data
x_test_cough = oe.fit_transform(x_test[['cough']])
x_train_cough
```

```python
# OneHotEncoding -> gender, city
ohe = OneHotEncoder(drop= 'first', dtype= 'int32', sparse_output= False)
x_train_gender_city = ohe.fit_transform(x_train[['gender','city']])

# also the test data
x_test_gender_city = ohe.fit_transform(x_test[['gender','city']])
x_train_gender_city.shape
```

```python
# extracting age

x_train_age = x_train[['age']]
x_test_age = x_test[['age']]

x_train_age.shape
```

```python
x_train_transformed = np.concatenate((x_train_age, x_train_fever, x_train_cough, x_train_gender_city), axis=1)
x_test_transformed = np.concatenate((x_test_age, x_test_fever, x_test_cough, x_test_gender_city), axis=1)
x_train_transformed.shape
```

# Mentos Zindagi -> with ColumnTransformer

```python
transformer = ColumnTransformer(transformers=[
    ('tnf1', SimpleImputer(),['fever']),
    ('tnf2',OrdinalEncoder(categories=[['Mild','Strong']]),['cough']),
    ('tnf3',OneHotEncoder(sparse_output=False,drop='first',dtype='int32'),['gender','city'])
    ], remainder= 'passthrough')
```

```python
transformed = transformer.fit_transform(x_train)
transformed.shape
```
