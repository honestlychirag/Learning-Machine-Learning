```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error, root_mean_squared_error
```

```python
df = pd.read_csv("E:/100 Days Machine Learning/Data Sets/placement_03.csv")
```

```python
plt.scatter(df['cgpa'],df['package']);
```

```python
x = df[['cgpa']]
y = df[['package']]
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.2, random_state=50)
```

```python
lr = LinearRegression()

lr.fit(x_train, y_train)
```

```python
y_pred = lr.predict(x_test)
```

```python
# MAE (LPA)
mean_absolute_error(y_test, y_pred)
```

```python
# MSE (LPA**2)

mean_squared_error(y_test, y_pred)
```

```python
# RMSE (LPA)

root_mean_squared_error(y_test, y_pred)
```

```python
# r2_score (cgpa is able to explain 86% variance in package)
r2_score(y_test, y_pred)
```

```python
y_test.shape[1]
```

```python
# adjusted r2_score

def adj_r2_score(y_test, y_pred):
    return 1 - ((1-r2_score(y_test, y_pred))*(y_test.shape[0]-1))/(y_test.shape[0] - 1 - y_test.shape[1])

adj_r2_score(y_test, y_pred)
```
