```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
from sklearn.datasets import make_regression
from sklearn.linear_model import LinearRegression
```

```python
x,y = make_regression(n_samples= 4, n_features= 1, n_informative= 1, n_targets= 1, noise= 100, random_state= 50)
```

```python
lr = LinearRegression()
lr.fit(x,y)
```

```python
lr.coef_
```

```python
lr.intercept_
```

```python
plt.plot(x,lr.predict(x), c='r')

plt.scatter(x,y);
```

1. Applying Gradient Descent.
2. Assuming : m = 76.27903802 --> (m is fixed)
3. let's take "initial C = 0"

```python
c = 100
m = 76.27903802
rate = 0.1

epochs = 10

for i in range(epochs):
    loss_slope = (-2) * np.sum(y - (m*x).ravel() - c)
    c = c - (rate*loss_slope)
    
    y_pred = m*x + c
    
    plt.plot(x,y_pred, c='g' ,alpha = 1-i/10)
    
plt.plot(x, lr.predict(x), 'r')
plt.scatter(x,y)
plt.show()
```
