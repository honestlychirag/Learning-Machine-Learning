```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv("E:/100 Days Machine Learning/Data Sets/titanic_02.csv")
df = df.rename(columns={"number":"family_members"})
```

Working on "family_members"

```python
df['family_members'].unique()
```

Method 1 : Convert 'A' (Alone) -> '0. Then convert all values to integer

```python
df['family_members'] = df['family_members'].replace('A','0').astype('int32')
```

Working on "Cabin"

```python
df['Cabin'].unique()
```

```python
df['cabin_no'] = df['Cabin'].str.extract('(\d+)')   # captures only numeric part of data
df['cabin_cat'] = df['Cabin'].str[0]                # captures first letter only
```
