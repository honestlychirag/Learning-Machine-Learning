# Normalization

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('Data Sets//wine_data.csv', usecols=[0,1,2])
df.columns = ['Class label', 'Alcohol', 'Malic acid']
df
```

```python
sns.kdeplot(df['Alcohol']);
```

```python
sns.kdeplot(df['Malic acid']);
```

```python
sns.scatterplot(data= df, x= 'Alcohol', y= 'Malic acid', hue= 'Class label', palette=['Red','Green','Blue']);
```

# Train Test Split

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(df.drop(columns= 'Class label'), df['Class label'], test_size= 0.3, random_state= 0)
```

```python
x_train.shape, x_test.shape
```

# Scaling time...

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
scaler.fit(x_train)
```

```python
x_train_scaled = scaler.transform(x_train)
x_test_scaled = scaler.transform(x_test)
```

```python
x_train_scaled = pd.DataFrame(x_train_scaled, columns= x_train.columns)
x_test_scaled = pd.DataFrame(x_test_scaled, columns= x_test.columns)
```

```python
x_train.describe(), x_train_scaled.describe()
```

```python
fig, (ax1,ax2) = plt.subplots(1,2, figsize= (12,5))

ax1.scatter(x_train['Alcohol'], x_train['Malic acid'], c= y_train)
ax1.set_title('Before Scaling')

ax2.scatter(x_train_scaled['Alcohol'], x_train_scaled['Malic acid'], c= y_train)
ax2.set_title('After Scaling')

plt.tight_layout()
plt.show()
```

```python
fig, (ax1,ax2) = plt.subplots(1,2, figsize= (12,5))

ax1.set_title('Before Scaling')
sns.kdeplot(x_train['Alcohol'], ax= ax1)
sns.kdeplot(x_train['Malic acid'], ax= ax1)
ax1.legend(['Alcohol','Malic acid'])

ax2.set_title('After Scaling')
sns.kdeplot(x_train_scaled['Alcohol'])
sns.kdeplot(x_train_scaled['Malic acid'])
ax2.legend(['Alcohol','Malic acid'])

plt.tight_layout()
plt.show()
```
