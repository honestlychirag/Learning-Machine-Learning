**Steps for end-to-end-ml-project :**

1. Preprocess + EDA + Feature Selection
2. Extract input & output cols
3. Train Test Split (TTS) & Scaling 
4. Model Training
5. Evaluate the model/model selection
6. Deployment

```python
# importing essential libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
from mlxtend.plotting import plot_decision_regions
import pickle
```

```python
placement = pd.read_csv("Data Sets/placement.csv")
placement
```

Step 1. Preprocess + EDA + Feature Selection

```python
# Preprocessing -> removing unwanted cols

placement = placement.drop(columns= "Unnamed: 0")
placement
```

```python
# EDA + Feature Selection
sns.relplot(kind='scatter', data=placement, x='cgpa', y='iq',hue= 'placement')
plt.show()
```

Step 2. Extracting input and output cols

```python
x = placement[['cgpa','iq']]    # input
y = placement['placement']      # output
```

Step 3. Train-Test-Split & Scaling

```python
x_train, x_test, y_train, y_test = train_test_split(x ,y , test_size= 0.1 )    # 0.1 => 90% data -> training, 10% data -> test
```

```python
# Scaling -> The process of adjusting the range and distribution of input variables (features) so they are on a similar scale.

scaler = StandardScaler()
x_train = scaler.fit_transform(x_train)
```

```python
x_test = scaler.transform(x_test)
```

Step 4. Model Training

```python
clf = LogisticRegression()
```

```python
clf.fit(x_train,y_train)
```

```python
y_pred = clf.predict(x_test)
print(y_pred)
print(y_test.values)
```

```python
# finding the accuracy of our model
accuracy_score(y_test, y_pred)
```

```python
# Plotting decision boundary
plot_decision_regions(x_train,y_train.values, clf= clf, legend= 2)
plt.show()
```

```python
# exporting our model

pickle.dump(clf, open('model.pkl','wb'))
```
